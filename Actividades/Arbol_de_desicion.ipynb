```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from IPython.display import Image
from sklearn import tree
from sklearn.datasets import load_breast_cancer
```

```python
# Cargar datos
data = load_breast_cancer()

# Caracteristicas y etiquetas
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python

# Exactitud del modelo en el entrenamiento.
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

```python
y_pred = clf.predict(X_test)
```

```python
# Evaluar exactitud
train_accuracy = clf.score(X_train, y_train)
print("Training Accuracy:", train_accuracy)

# Predecir etiquetar para prueba
y_pred = clf.predict(X_test)

# Calcular exactitud de prueba
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)

# generar reporte de clasificación
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python

plt.figure(figsize=(20, 15))

# Visualizar arbol de decisión
tree.plot_tree(clf, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.title("Decision Tree for Breast Cancer Classification")
plt.savefig("breast_cancer_decision_tree.png", dpi=300)  # Save the plot
plt.show()
```
